### 合併 txt

```python
import pandas as pd 
import os
path = "Train/dataTrainComplete"
files = os.listdir(path)
txts = []
for file in files:
    position = path + '\\' + file
    with open(position, "r", encoding = 'utf-8') as f:
        data = f.read()
        ndf = pd.DataFrame([{'Num':file.split('.txt')[0],
                             'Content':data}],
                            columns = ['Num', 'Content'])
        txts.append(ndf)
```

```python
test = pd.concat(txts, ignore_index=True)
test.to_csv('stage1_origin.csv', encoding='utf_8_sig')
```

### Jaccard Similarity

```python
import pandas as pd 
stage1_unigram = pd.read_csv('stage1_unigram.csv', encoding='utf_8_sig') 
stage1_unigram
```

```python
article_index = [str(i) for i in list(stage1_unigram.Num)]
article_index[:10]
```

```python
def jaccard_similarity(list1, list2):
    intersection = len(list(set(list1).intersection(list2)))
    union = (len(set(list1)) + len(set(list2))) - intersection
    return float(intersection) / union
```

```python
result_df = pd.DataFrame({'Test':[], 'Reference': [], 'Similarity(%)': []})
count = 0
similarity_list=[]
article_index[0]
for i in data_words_trigrams:
    for j in range(len(data_words_trigrams)):
        if i == data_words_trigrams[j]:
            continue
        similarity = jaccard_similarity(i, data_words_trigrams[j])
        dict = {'Test': article_index[count], 'Reference': article_index[j], 'Similarity(%)': similarity}
        result_df = result_df.append(dict, ignore_index=True)
    count+=1
```
