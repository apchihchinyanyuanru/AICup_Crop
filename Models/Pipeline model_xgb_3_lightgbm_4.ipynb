```python
import pandas as pd
import numpy as np
```

```python
# read S3 feature data
X3_similarity_of_times = np.load('../FeatureExtracting/stage3_similarity_of_only_times.npy')
X3_times = np.load('../FeatureExtracting/stage3_times_feature.npy')
```

```python
X3_times = np.delete(X3_times, 1529, 1)
X3_times = np.delete(X3_times, 764, 1)
```

```python
# read in feature data
X3 = np.concatenate((X3_similarity_of_times, X3_times), axis=1)
X3.shape
```

```python
X3[0]
```

```python
# read in feature data
features3_csv = pd.read_csv("../FeatureExtracting/stage3_similarity_features.csv")
```

```python
features3_csv
```

```python
# take only article pair orders
article_pairs = features3_csv[["Test", "Reference"]]
article_pairs
```

```python
# turn X3 features to frame
X3_df = pd.DataFrame(X3)
X3_df
```

```python
# join the article times & similarity times features with article pairs
joined_df = pd.merge(article_pairs, X3_df, left_index=True, right_index=True)
joined_df
```

```python
# read in the best results
xgb_df = pd.read_csv("csv_results/3_XGBOOST.csv")
xgb_df
```

```python
# join back features from the best csv result
final_df = pd.merge(xgb_df, joined_df, on=["Test", "Reference"], how="left")
final_df
```

```python
# get the numpy features of the best results 
X3_bests = final_df.copy()
del X3_bests["Test"]
del X3_bests["Reference"]
X3_bests_np = X3_bests.to_numpy()
X3_bests_np
```

```python
len(X3_bests_np[0])
```

```python
len(X3_bests_np)
```

```python
# write to a numpy file
with open("X3_bests.npy", "wb") as f:
    np.save(f, X3_bests_np)
```

```python
# read back the lightgbm results and change to dataframe
lgb_np = np.load("Y3_best.npy")
lgb_df = pd.DataFrame(lgb_np)
lgb_df = lgb_df.rename({0: 'Relevance'}, axis=1)
lgb_df
```

```python
# join back lightgbm results with xgboost results & do the filtem
best_pairs = final_df[["Test", "Reference"]]
pipe_df = pd.merge(best_pairs, lgb_df, left_index=True, right_index=True)
pipe_df
```

```python
pipe_df.loc[pipe_df['Relevance'] == 1][['Test', 'Reference']].to_csv("csv_results/xgb3_lgbm4_pipe.csv", index=False)
```
