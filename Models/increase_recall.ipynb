```python
# load package
import pandas as pd
import numpy as np
```

```python
lost_2 = pd.read_csv("/content/drive/MyDrive/ISA_DM/Result/lgbm_lost_model2.csv")
lost_3 = pd.read_csv("/content/drive/MyDrive/ISA_DM/Result/lgbm_lost_model3.csv")
lost_4 = pd.read_csv("/content/drive/MyDrive/ISA_DM/Result/lgbm_lost_model4.csv")
lost_5 = pd.read_csv("/content/drive/MyDrive/ISA_DM/Result/lgbm_lost_model5.csv")
lost_8 = pd.read_csv("/content/drive/MyDrive/ISA_DM/Result/lgbm_lost_model8.csv")
lost_9 = pd.read_csv("/content/drive/MyDrive/ISA_DM/Result/lgbm_lost_model9.csv")
all_df = pd.concat([lost_2, lost_3, lost_4, lost_5, lost_8, lost_9])
all_df
```

```python
all_df['Amount'] = 1
all_df['Total'] = all_df.groupby(["Test",	"Reference","Similarity"])['Amount'].transform('sum')
all_df
```

```python
del all_df['Amount']
```

```python
all_df = all_df.drop_duplicates() # 他group by之後還是都存在 問號 :) ? 所以才把重複的drop調
```
