```python
# load package
import pandas as pd
import numpy as np
```

```python
# read data
df_model = pd.read_csv('../../input/ReallyBig.csv', header = 0)
# df_model = df_model[0:5]
df_model
```

```python
# create X values
X_df = df_model.copy()
del X_df['Test']
del X_df['Reference']
del X_df['Relevance']
X_df
```

```python
# turn X_df into numpy array
X = X_df.values
X
```

```python
# get labels for classifier
Y = df_model.Relevance.values

# encode labels
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(Y)
y
```

```python
# split data into testing set and training set
# 80% of data into the training set, while the rest 20% is used as a testing set.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=345)
```

```python
# train model
from lightgbm import LGBMClassifier
from sklearn.utils import class_weight

train_sample = class_weight.compute_sample_weight('balanced', y_train)
test_sample = class_weight.compute_sample_weight('balanced', y_test)

gbm = LGBMClassifier(num_leaves=31, learning_rate=0.05, n_estimators=40)
# gbm.fit(X_train, y_train, eval_set=[(X_test, y_test)], early_stopping_rounds=5,  sample_weight=train_sample)
gbm.fit(X_train, y_train, eval_set=[(X_test, y_test)], early_stopping_rounds=5)
```

```python
# 模型存储
import joblib

joblib.dump(gbm, 'loan_model.pkl')
# 模型加载
gbm = joblib.load('loan_model.pkl')
```

```python
# 模型预测
y_pred = gbm.predict(X_test, num_iteration=gbm.best_iteration_)
```

```python
# 模型评估
from sklearn.metrics import accuracy_score

print('The accuracy of prediction is:', accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_true=y_test, y_pred=y_pred, digits=4))
```

```python
# 特征重要度
print('Feature importances:', list(gbm.feature_importances_))
```

```python
# 网格搜索，参数优化
from sklearn.model_selection import GridSearchCV

estimator = LGBMClassifier(num_leaves=31)
param_grid = {
    'learning_rate': [0.01, 0.05, 0.1, 1],
    'n_estimators': [20, 40]
}
gbm = GridSearchCV(estimator, param_grid)
gbm.fit(X_train, y_train)
print('Best parameters found by grid search are:', gbm.best_params_)
```
