```python
# load package
import pandas as pd
import numpy as np
```

```python
Y3_2 = np.load('/content/drive/MyDrive/ISA_DM/Result/final/Y3_xgb3_lgbm2.npy')
Y3_3 = np.load('/content/drive/MyDrive/ISA_DM/Result/final/Y3_xgb3_lgbm3.npy')
Y3_4 = np.load('/content/drive/MyDrive/ISA_DM/Result/final/Y3_xgb3_lgbm4.npy')
Y3_5 = np.load('/content/drive/MyDrive/ISA_DM/Result/final/Y3_xgb3_lgbm5.npy')
Y3_8 = np.load('/content/drive/MyDrive/ISA_DM/Result/final/Y3_xgb3_lgbm8.npy')
Y3_9 = np.load('/content/drive/MyDrive/ISA_DM/Result/final/Y3_xgb3_lgbm9.npy')
Y3_rf4 = np.load('/content/drive/MyDrive/ISA_DM/Result/final/Y3_xgb3_rf4.npy')
Y3_rf3 = np.load('/content/drive/MyDrive/ISA_DM/Result/final/Y3_xgb3_rf3.npy')
```

```python
# read in stage 2 labels for final double check

res_df = pd.read_csv('/content/drive/MyDrive/ISA_DM/Result/final/3_XGBOOST.csv')
res_df['lgbm2'] = Y3_2
res_df['lgbm3'] = Y3_3
res_df['lgbm4'] = Y3_4
res_df['lgbm5'] = Y3_5
res_df['lgbm8'] = Y3_8
res_df['lgbm9'] = Y3_9
res_df['rf4'] = Y3_rf4
res_df['rf3'] = Y3_rf3

res_df
```

```python
res_df_only_1 = res_df.loc[res_df['lgbm5'] == 1]
res_df_only_1
```

```python
res_df['Total'] = res_df[['lgbm2', 'lgbm3', 'lgbm4', 'lgbm5', 'lgbm8', 'lgbm9', 'rf4', 'rf3']].sum(axis=1)
```

```python
res_df.loc[res_df.Total > 0][["Test", "Reference"]].to_csv("/content/drive/MyDrive/ISA_DM/Result/xgb_lgbm_rf34_full_0.csv", index=False)
```

```python
res_df.loc[res_df.Total > 0]
```

```python
res_df.loc['Model_sum'] = res_df.apply(lambda x: x.sum())
```

```python
# left join label similarity
simi_template = pd.read_csv('/content/drive/MyDrive/ISA_DM/Result/model feed in features/result_stage3_template.csv')
res_df = pd.merge(res_df,simi_template,on=('Test','Reference'),how='left')
```

```python
res_df.loc[(res_df.Similarity >= 0.6) & (res_df.Total > 0)][["Test", "Reference"]].to_csv("/content/drive/MyDrive/ISA_DM/Result/xgb_lgbm_rf34_60_0.csv", index=False)
```
